1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
data = pd.read_csv(url)

data.columns = data.columns.str.lower().str.replace(' ', '_')

filtered_data = data[(data['total_claim_amount'] > 1000) & (data['response'] == 'Yes')]

display(filtered_data)
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
claim_by_policy_gender = data[data['response'] == 'Yes'].pivot_table(values='total_claim_amount', index=['policy_type', 'gender'], aggfunc='mean')

display(claim_by_policy_gender)
```

There is not a major discrepancy between male and female, but with a small trending of female earning more.
The policy Personal Auto is the one that have the highiest amount of claims.

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
customers_by_state = data['state'].value_counts()
customers_by_state_filtered = customers_by_state[customers_by_state > 500]

display(customers_by_state_filtered)
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
stats_by_education_gender = data.pivot_table(values='customer_lifetime_value', index=['education', 'gender'], aggfunc={'max', 'min', 'median'})

display(stats_by_education_gender)
```

Female bachelor has access to a higher amount than male. This could be due the age of bachelor male tend to commit more accidents.

In college woman have a higher max than male, but when looking in median male has a higher value, which indicate that there is another layer being considered apart of the education that is taking a key aspect to determine the number.

In Doctor there is a discrepency between max for male and female being female the higher paid one, but looking to median it's similar. This indicate that have less female with doctor eduction, which could lead to the increase of the max.
